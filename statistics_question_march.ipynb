 What is the meaning of six sigma in statistics?  Give proper example


Six Sigma is a set of methodologies and tools used to improve business processes by reducing defects and errors, minimizing variation, and increasing quality and efficiency. The goal of Six Sigma is to achieve a level of quality that is nearly perfect, with only 3.4 defects per million opportunities. This is achieved by using a structured approach called DMAIC (Define, Measure, Analyze, Improve, Control) to identify and eliminate causes of variation and improve processes.

example: Six Sigma in Online Order Processing

Imagine an e-commerce company that aims to achieve Six Sigma quality in processing online orders. One critical quality characteristic is the time it takes to process an order from the moment it's placed until it's ready for shipment.

Mean (μ): The target processing time for an order is 24 hours.

Standard Deviation (σ): In a Six Sigma process, the standard deviation is minimized, indicating that the processing times are very consistent.

Six Sigma Level: The company aims for a Six Sigma level, meaning that the acceptable range (±3σ) covers almost all orders processed.

If the standard deviation is, for example, 2 hours, it means that 99.73% of the orders will be processed within the range of 18 to 30 hours. This high level of precision ensures that the majority of orders are processed within a predictable time frame.

What type of data does not have a log-normal distribution or a Gaussian distribution?  Give proper example

ans - Data that does not follow  a log-normal or Gaussian distribution often includes instances of heavy-tailed distributions, such as power-law distributions.
Example: Income distribution

In a power-law distribution of income, a small proportion of the population earns a disproportionately large share of the total income. Most people have relatively low income, while a few individuals or entities have very high income. This distribution does not conform to a normal or log-normal distribution because it lacks the symmetry and tends to have a heavy tail.

```python
'''
Q3 What is the meaning of the five-number summary in Statistics? Give proper example

A summary consists of five values: the most extreme values in the data set (the maximum and minimum values), the lower and upper quartiles, and the median. These values are presented together and ordered from lowest to highest: minimum value, lower quartile (Q1), median value (Q2), upper quartile (Q3), maximum value.
A five-number summary can be represented in a diagram known as a box and whisker plot.

Minimum: The smallest value in the dataset.
First Quartile (Q1): The value below which 25% of the data falls. It is the median of the lower half of the dataset.
Median (Q2): The middle value of the dataset when it is sorted in ascending order. It separates the lower 50% from the upper 50% of the data.
Third Quartile (Q3): The value below which 75% of the data falls. It is the median of the upper half of the dataset.
Maximum: The largest value in the dataset.

example:-
Consider the following dataset representing the number of hours per week spent on studying by a group of students:

2,4,5,6,7,8,8,9,10,12,15,18,20

Minimum: 2 (the smallest value in the dataset).
First Quartile (Q1): 6 (the median of the lower half: 2, 4, 5, 6, 7).
Median (Q2): 8 (the middle value when the dataset is sorted: 2, 4, 5, 6, 7, 8, 8, 9, 10, 12, 15, 18, 20).
Third Quartile (Q3): 12 (the median of the upper half: 8, 9, 10, 12, 15, 18, 20).
Maximum: 20 (the largest value in the dataset).
'''

## example of boxplot 
!pip install matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# Example dataset
data = [2, 4, 5, 6, 7, 8, 8, 9, 10, 12, 15, 18, 20]

# Create a box plot
plt.boxplot(data, vert=False)

# Add labels
plt.title("Box Plot with Five-Number Summary")
plt.xlabel("Study Hours per Week")

# Display the plot
plt.show()
```

```python
'''
Q4 What is correlation? Give an example with a dataset & graphical representation on jupyter Notebook

Correlation refers to the statistical relationship between the two entities. It measures the extent to which two variables are linearly related. For example, the height and weight of a person are related, and taller people tend to be heavier than shorter people.
correlation helps us understand how changes in one variable might be associated with changes in another variable.
For example, the height and weight of a person are related, and taller people tend to be heavier than shorter people

there are three types of correlation:

1 indicates a perfect positive correlation: as one variable increases, the other variable increases proportionally.
-1 indicates a perfect negative correlation: as one variable increases, the other variable decreases proportionally.
0 indicates no linear correlation: there is no systematic linear relationship between the two variables.
Correlation can be calculated using various methods, with the Pearson correlation coefficient being the most common for measuring linear relationships. Other methods include Spearman rank correlation and Kendall's tau, which are used when dealing with non-linear relationships or ordinal data.
'''
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a synthetic dataset
np.random.seed(42)
x = np.random.rand(50) * 10  # Independent variable
y = 2 * x + 1 + np.random.randn(50) * 2  # Dependent variable with some noise

# Create a DataFrame
df = pd.DataFrame({'X': x, 'Y': y})

# Calculate the correlation coefficient
correlation_coefficient = df['X'].corr(df['Y'])

# Plot the scatter plot and regression line
plt.figure(figsize=(8, 6))
sns.scatterplot(x='X', y='Y', data=df, color='blue', label=f'Correlation: {correlation_coefficient:.2f}')
sns.regplot(x='X', y='Y', data=df, color='red')

# Add labels and title
plt.title('Scatter Plot with Regression Line')
plt.xlabel('X')
plt.ylabel('Y')

# Display the plot
plt.legend()
plt.show()

```
